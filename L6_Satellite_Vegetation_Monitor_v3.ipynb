### Script 1 — Seasonal NDVI Monitor

````python
script_1 = '''/* ════════════════════════════════════════════════════════════════════════
   SCRIPT 1 of 5 — QUICKSTART: Seasonal NDVI Monitor (with custom AOI)
   AI for Sustainable Development · Lesson 6 · ESDU @ AUB · karianet.org
   ════════════════════════════════════════════════════════════════════════
   WHAT THIS SCRIPT DOES
   Loads Sentinel-2 imagery for your chosen area across 4 seasons,
   calculates NDVI (vegetation health), and produces a map + chart.
   This is the same core script from before — now rebuilt so you can
   plug in ANY area of interest using one of 4 different methods.
   See "AOI_Guide.md" for full instructions on each method.
   ════════════════════════════════════════════════════════════════════════ */


/* ──────────────────────────────────────────────────────────────────────
   STEP 1 — CHOOSE YOUR AREA OF INTEREST (AOI)
   ──────────────────────────────────────────────────────────────────────
   Pick ONE of the four methods below by uncommenting it (remove the //
   at the start of the line) and commenting out the others.
   Full explanation of each method is in the AOI_Guide.md document.
   ────────────────────────────────────────────────────────────────────── */

// ── METHOD A: Use a preset location from our location list ──────────────
// Copy any rectangle from the "PRESET LOCATIONS" list at the bottom of
// this script and paste it here. Default shown: Bekaa Valley.
var aoi = ee.Geometry.Rectangle([35.85, 33.55, 36.30, 33.90]);   // Bekaa Valley

// ── METHOD B: Draw your own polygon (see Step 1B below) ──────────────────
// var aoi = geometry;   // uncomment this line AFTER drawing — see Step 1B

// ── METHOD C: Search a place by name ──────────────────────────────────────
// Type any place name into the search bar at the TOP of the Code Editor
// map (magnifying glass icon), press Enter, then run:
// var aoi = ee.FeatureCollection('FAO/GAUL/2015/level2')
//   .filter(ee.Filter.eq('ADM2_NAME', 'Zahle')).geometry();   // example: Zahle district

// ── METHOD D: Point + buffer radius (a circle around one coordinate) ─────
// var aoi = ee.Geometry.Point([35.90, 33.85]).buffer(5000);   // 5km radius around Zahle


/* ──────────────────────────────────────────────────────────────────────
   STEP 1B — HOW TO DRAW YOUR OWN POLYGON (Method B)
   ──────────────────────────────────────────────────────────────────────
   1. Look at the TOP-LEFT of the map panel (below this code, not above)
   2. Click the small polygon icon (looks like a pentagon shape)
   3. Click on the map to place each corner of your shape
   4. Double-click to finish the polygon
   5. A new section called "Geometry Imports" appears ABOVE this script
   6. Earth Engine automatically creates a variable called "geometry"
   7. Go back up to STEP 1 above, comment out Method A,
      and uncomment the line: var aoi = geometry;
   8. Click Run again
   ────────────────────────────────────────────────────────────────────── */

Map.centerObject(aoi, 9);
Map.addLayer(aoi, {color: 'red'}, 'My Area of Interest');
print('✅ STEP 1 complete. Area selected — check the MAP for a red outline.');
print('   Area size: approx', aoi.area().divide(1000*1000).round(), 'km²');


/* ──────────────────────────────────────────────────────────────────────
   STEP 2 — SET YOUR YEAR
   ────────────────────────────────────────────────────────────────────── */
var YEAR = 2023;   // ✏️ change to any year from 2017 to 2026


/* ──────────────────────────────────────────────────────────────────────
   STEP 3 — LOAD CLOUD-FREE SENTINEL-2 IMAGERY (reusable function)
   ──────────────────────────────────────────────────────────────────────
   Sentinel-2 is a free satellite from the European Space Agency.
   Two satellites orbit together and photograph nearly every point on
   Earth every 5 days, at 10-metre resolution. We filter out images
   with more than 20% cloud cover, then take the median of what's left
   — this naturally removes any leftover clouds through averaging.
   ────────────────────────────────────────────────────────────────────── */
function getCloudFreeImage(startDate, endDate) {
  var collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate(startDate, endDate)
    .filterBounds(aoi)
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
    .map(function(img) { return img.divide(10000); });
  return collection.median().clip(aoi);
}


/* ──────────────────────────────────────────────────────────────────────
   STEP 4 — CALCULATE NDVI
   ──────────────────────────────────────────────────────────────────────
   NDVI = (Near-Infrared − Red) / (Near-Infrared + Red)
   Healthy plants reflect NIR strongly and absorb Red for photosynthesis,
   so NDVI close to +1 = lush vegetation, close to 0 or negative = bare
   soil, water, or built surfaces.
   ────────────────────────────────────────────────────────────────────── */
function addNDVI(image) {
  return image.addBands(image.normalizedDifference(['B8', 'B4']).rename('NDVI'));
}


/* ──────────────────────────────────────────────────────────────────────
   STEP 5 — LOAD 4 SEASONS, ADD TO MAP
   ────────────────────────────────────────────────────────────────────── */
var seasons = [
  {name: 'Winter', start: YEAR + '-01-01', end: YEAR + '-03-31'},
  {name: 'Spring', start: YEAR + '-04-01', end: YEAR + '-06-30'},
  {name: 'Summer', start: YEAR + '-07-01', end: YEAR + '-09-30'},
  {name: 'Autumn', start: YEAR + '-10-01', end: YEAR + '-12-31'}
];

var ndviVis = {min: -0.2, max: 0.8, palette: ['red','orange','yellow','lightgreen','darkgreen']};
var seasonalNDVI = {};

seasons.forEach(function(s) {
  var img = addNDVI(getCloudFreeImage(s.start, s.end)).select('NDVI');
  seasonalNDVI[s.name] = img;
  Map.addLayer(img, ndviVis, 'NDVI - ' + s.name, false);
});

print('✅ STEP 5 complete — 4 seasonal NDVI layers added.');
print('   Tick each layer checkbox (top-right of MAP) to compare seasons.');


/* ──────────────────────────────────────────────────────────────────────
   STEP 6 — CHART + PRINTED RESULTS
   ────────────────────────────────────────────────────────────────────── */
var chartData = seasons.map(function(s) {
  var meanDict = seasonalNDVI[s.name].reduceRegion({
    reducer: ee.Reducer.mean(), geometry: aoi, scale: 100, maxPixels: 1e9
  });
  return ee.Feature(null, {season: s.name, NDVI: meanDict.get('NDVI')});
});

print(ui.Chart.feature.byFeature(ee.FeatureCollection(chartData), 'season', ['NDVI'])
  .setChartType('LineChart')
  .setOptions({title: 'Mean NDVI by Season — ' + YEAR, colors: ['2e7d32'], lineWidth: 3, pointSize: 7}));

print('═══ YOUR RESULTS (copy into your notebook) ═══');
seasons.forEach(function(s) {
  seasonalNDVI[s.name].reduceRegion({
    reducer: ee.Reducer.mean(), geometry: aoi, scale: 100, maxPixels: 1e9
  }).get('NDVI').evaluate(function(v) { print(s.name + ' mean NDVI: ' + v.toFixed(3)); });
});


/* ════════════════════════════════════════════════════════════════════════
   PRESET LOCATIONS — copy any line into Step 1, Method A
   ════════════════════════════════════════════════════════════════════════
   Bekaa Valley, Lebanon   : ee.Geometry.Rectangle([35.85, 33.55, 36.30, 33.90])
   South Lebanon farmland  : ee.Geometry.Rectangle([35.35, 33.05, 35.65, 33.30])
   Akkar plain, Lebanon    : ee.Geometry.Rectangle([35.95, 34.45, 36.25, 34.65])
   Nile Delta, Egypt       : ee.Geometry.Rectangle([30.50, 30.60, 31.50, 31.40])
   Jordan Valley           : ee.Geometry.Rectangle([35.50, 31.85, 35.65, 32.60])
   Beirut urban area       : ee.Geometry.Rectangle([35.45, 33.85, 35.55, 33.92])
   ════════════════════════════════════════════════════════════════════════ */
'''
print("Script 1 loaded. Run the display cell below to copy it.")
````

### Script 2 — False Color Vision

````python
script_2 = '''/* ════════════════════════════════════════════════════════════════════════
   SCRIPT 2 of 5 — FALSE COLOR VISION: See Vegetation Health Instantly
   AI for Sustainable Development · Lesson 6 · ESDU @ AUB · karianet.org
   ════════════════════════════════════════════════════════════════════════
   WHAT THIS SCRIPT DOES
   Displays your area in "False Color Infrared" — a technique that
   reassigns satellite light bands to colors your eyes can see, making
   healthy vegetation glow BRIGHT RED and revealing detail invisible
   in a normal photo. This is the single most visually striking and
   most scientifically standard technique in remote sensing — used in
   NASA and ESA imagery you've likely seen in documentaries.

   WHY IT WORKS
   Your eyes can only see 3 colors: Red, Green, Blue (RGB). Satellites
   capture light far beyond that — including Near-Infrared (NIR), a
   wavelength healthy plants reflect VERY strongly. By feeding the NIR
   band into the "Red" channel of your screen, anything that reflects
   a lot of NIR (healthy vegetation) appears as a glowing red color —
   even though no actual red light is involved. This isn't a trick or
   an exaggeration — it's a direct, faithful representation of real
   physical reflectance data, just translated into colors we can see.
   ════════════════════════════════════════════════════════════════════════ */


/* STEP 1 — Choose your area (see Script 1 or AOI_Guide.md for all methods) */
var aoi = ee.Geometry.Rectangle([35.85, 33.55, 36.30, 33.90]);   // Bekaa Valley
// To draw your own: use the polygon tool top-left of the map, then:
// var aoi = geometry;

Map.centerObject(aoi, 10);


/* STEP 2 — Load a single cloud-free composite for the period you want */
var YEAR = 2023;
var START = YEAR + '-06-01';   // ✏️ change date range as needed
var END   = YEAR + '-08-31';

var image = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
  .filterDate(START, END)
  .filterBounds(aoi)
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 15))
  .median()
  .divide(10000)
  .clip(aoi);

print('✅ STEP 2 complete — composite loaded for', START, 'to', END);


/* ──────────────────────────────────────────────────────────────────────
   STEP 3 — TRUE COLOR (what your eyes would actually see)
   ──────────────────────────────────────────────────────────────────────
   Bands: B4 (Red), B3 (Green), B2 (Blue) — assigned to Red, Green, Blue
   This is the "normal photo" baseline to compare against.
   ────────────────────────────────────────────────────────────────────── */
var trueColorVis = {bands: ['B4', 'B3', 'B2'], min: 0, max: 0.3};
Map.addLayer(image, trueColorVis, '1. True Color (normal photo)');


/* ──────────────────────────────────────────────────────────────────────
   STEP 4 — FALSE COLOR INFRARED (vegetation glows red)
   ──────────────────────────────────────────────────────────────────────
   Bands: B8 (Near-Infrared), B4 (Red), B3 (Green) — assigned to R, G, B
   Healthy vegetation → bright red. Water → black/dark blue.
   Bare soil/urban → white, tan, or grey.
   ────────────────────────────────────────────────────────────────────── */
var falseColorVis = {bands: ['B8', 'B4', 'B3'], min: 0, max: 0.4};
Map.addLayer(image, falseColorVis, '2. False Color Infrared (vegetation = red)');


/* ──────────────────────────────────────────────────────────────────────
   STEP 5 — AGRICULTURE COMPOSITE (best for distinguishing crop types)
   ──────────────────────────────────────────────────────────────────────
   Bands: B11 (Short-wave Infrared), B8 (NIR), B2 (Blue)
   Dense healthy crops appear dark green. Useful for telling apart
   different crop types and irrigation status side-by-side.
   ────────────────────────────────────────────────────────────────────── */
var agricultureVis = {bands: ['B11', 'B8', 'B2'], min: 0, max: 0.4};
Map.addLayer(image, agricultureVis, '3. Agriculture Composite (crop detail)');


/* ──────────────────────────────────────────────────────────────────────
   STEP 6 — URBAN FALSE COLOR (built-up areas in white/grey/purple)
   ──────────────────────────────────────────────────────────────────────
   Bands: B12 (SWIR2), B11 (SWIR1), B4 (Red)
   Buildings, roads, and concrete appear white/grey/purple.
   Vegetation appears green. Water appears black/dark blue.
   Great for seeing urban sprawl next to farmland at a glance.
   ────────────────────────────────────────────────────────────────────── */
var urbanVis = {bands: ['B12', 'B11', 'B4'], min: 0, max: 0.4};
Map.addLayer(image, urbanVis, '4. Urban False Color (buildings = white/purple)');


print('✅ ALL 4 LAYERS LOADED.');
print('   Use the LAYERS icon (top-right of the map) to toggle each one ON/OFF.');
print('   Try switching between Layer 1 and Layer 2 repeatedly —');
print('   notice how vegetation suddenly "lights up" red in Layer 2.');
print('   This same NIR reflectance is exactly what NDVI measures numerically');
print('   in Script 1 — here you are seeing the SAME physical phenomenon,');
print('   just as a picture instead of a number.');


/* ════════════════════════════════════════════════════════════════════════
   REFLECTION PROMPT (answer this in your notebook)
   Compare Layer 1 (True Color) and Layer 4 (Urban False Color) for your
   area. Can you find a boundary between a city/town and farmland? How
   clearly does each composite show that boundary, and why do you think
   that is, based on what each band combination is designed to detect?
   ════════════════════════════════════════════════════════════════════════ */
'''
print("Script 2 loaded. Run the display cell below to copy it.")
````

### Script 3 — Time Machine (Change Detection)

````python
script_3 = '''/* ════════════════════════════════════════════════════════════════════════
   SCRIPT 3 of 5 — TIME MACHINE: Detect Change Between Two Years
   AI for Sustainable Development · Lesson 6 · ESDU @ AUB · karianet.org
   ════════════════════════════════════════════════════════════════════════
   WHAT THIS SCRIPT DOES
   Compares satellite imagery from two different years over the SAME
   area and highlights exactly where vegetation increased, decreased,
   or stayed the same. This is real change-detection — the same
   fundamental technique used to track deforestation, urban sprawl,
   drought impact, and agricultural expansion at a national scale.

   THE OUTPUT
   A single map where:
     RED    = vegetation LOST between your two years (concerning)
     GREEN  = vegetation GAINED between your two years (positive)
     GREY   = no significant change
   ════════════════════════════════════════════════════════════════════════ */


/* STEP 1 — Choose your area */
var aoi = ee.Geometry.Rectangle([35.85, 33.55, 36.30, 33.90]);   // Bekaa Valley
// To draw your own polygon: use the tool top-left of the map, then:
// var aoi = geometry;

Map.centerObject(aoi, 10);


/* STEP 2 — Choose your two comparison years */
var YEAR_BEFORE = 2018;   // ✏️ earlier year
var YEAR_AFTER  = 2023;   // ✏️ more recent year
var MONTH_START = 6;      // ✏️ same month range used for both years
var MONTH_END   = 8;      // (keeping season constant makes the comparison fair)


/* ──────────────────────────────────────────────────────────────────────
   STEP 3 — LOAD CLOUD-FREE COMPOSITES FOR BOTH YEARS
   ──────────────────────────────────────────────────────────────────────
   We use the SAME months in both years (e.g. both summers) so that
   any difference we detect is due to real change, not seasonal cycles.
   ────────────────────────────────────────────────────────────────────── */
function loadComposite(year) {
  var start = year + '-0' + MONTH_START + '-01';
  var end   = year + '-0' + MONTH_END + '-28';
  return ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate(start, end)
    .filterBounds(aoi)
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
    .median()
    .divide(10000)
    .clip(aoi);
}

var imageBefore = loadComposite(YEAR_BEFORE);
var imageAfter  = loadComposite(YEAR_AFTER);

print('✅ STEP 3 complete — loaded composites for', YEAR_BEFORE, 'and', YEAR_AFTER);


/* STEP 4 — Calculate NDVI for both years */
function ndvi(img) {
  return img.normalizedDifference(['B8', 'B4']).rename('NDVI');
}

var ndviBefore = ndvi(imageBefore);
var ndviAfter  = ndvi(imageAfter);


/* ──────────────────────────────────────────────────────────────────────
   STEP 5 — CALCULATE THE DIFFERENCE (this is the "change" itself)
   ──────────────────────────────────────────────────────────────────────
   Subtracting one NDVI map from another gives a "delta" map:
     Positive value → NDVI went UP  → vegetation increased (greening)
     Negative value → NDVI went DOWN → vegetation decreased (degradation)
     Near zero       → no meaningful change
   ────────────────────────────────────────────────────────────────────── */
var ndviChange = ndviAfter.subtract(ndviBefore).rename('NDVI_change');

var changeVis = {
  min: -0.3, max: 0.3,
  palette: ['8B0000', 'FF6347', 'FFFFFF', '90EE90', '006400']
  // dark red = big loss · light red = some loss · white = no change
  // light green = some gain · dark green = big gain
};

Map.addLayer(ndviBefore, {min:-0.2,max:0.8,palette:['red','yellow','green']}, '1. NDVI - ' + YEAR_BEFORE, false);
Map.addLayer(ndviAfter,  {min:-0.2,max:0.8,palette:['red','yellow','green']}, '2. NDVI - ' + YEAR_AFTER, false);
Map.addLayer(ndviChange, changeVis, '3. CHANGE MAP (red=loss, green=gain)', true);

print('✅ STEP 5 complete — change map added (visible by default).');
print('   Red areas = vegetation declined between', YEAR_BEFORE, 'and', YEAR_AFTER);
print('   Green areas = vegetation increased in that same period');


/* ──────────────────────────────────────────────────────────────────────
   STEP 6 — QUANTIFY THE CHANGE (turn the picture into real numbers)
   ────────────────────────────────────────────────────────────────────── */
var meanBefore = ndviBefore.reduceRegion({reducer: ee.Reducer.mean(), geometry: aoi, scale: 100, maxPixels: 1e9});
var meanAfter  = ndviAfter.reduceRegion({reducer: ee.Reducer.mean(), geometry: aoi, scale: 100, maxPixels: 1e9});

// Calculate the percentage of the area that lost vs gained vegetation
var lossArea = ndviChange.lt(-0.1).multiply(ee.Image.pixelArea());
var gainArea = ndviChange.gt(0.1).multiply(ee.Image.pixelArea());

var lossStats = lossArea.reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 100, maxPixels: 1e9});
var gainStats = gainArea.reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 100, maxPixels: 1e9});
var totalAreaM2 = aoi.area();

print('═══ YOUR RESULTS (copy into your notebook) ═══');
meanBefore.get('NDVI').evaluate(function(v) { print('Mean NDVI ' + YEAR_BEFORE + ': ' + v.toFixed(3)); });
meanAfter.get('NDVI').evaluate(function(v) { print('Mean NDVI ' + YEAR_AFTER + ': ' + v.toFixed(3)); });

lossStats.get('NDVI_change').evaluate(function(lossM2) {
  totalAreaM2.evaluate(function(totalM2) {
    var pct = (lossM2 / totalM2 * 100).toFixed(1);
    print('Area with vegetation LOSS: ' + pct + '% of total area');
  });
});
gainStats.get('NDVI_change').evaluate(function(gainM2) {
  totalAreaM2.evaluate(function(totalM2) {
    var pct = (gainM2 / totalM2 * 100).toFixed(1);
    print('Area with vegetation GAIN: ' + pct + '% of total area');
  });
});


/* ════════════════════════════════════════════════════════════════════════
   REFLECTION PROMPT (answer this in your notebook)
   What does the change map show for your chosen area? If you see large
   red zones, what real-world causes could explain vegetation loss over
   this time period (construction, drought, land-use change, conflict
   damage, deforestation)? If you see green zones, what could explain
   vegetation gain (reforestation, new irrigation, land reclamation)?
   You don't need certainty — propose a plausible hypothesis and explain
   what additional data would help you confirm it.
   ════════════════════════════════════════════════════════════════════════ */
'''
print("Script 3 loaded. Run the display cell below to copy it.")
````

### Script 4 — Water & Drought Monitor

````python
script_4 = '''/* ════════════════════════════════════════════════════════════════════════
   SCRIPT 4 of 5 — WATER & DROUGHT MONITOR
   AI for Sustainable Development · Lesson 6 · ESDU @ AUB · karianet.org
   ════════════════════════════════════════════════════════════════════════
   WHAT THIS SCRIPT DOES
   Maps surface water bodies (rivers, lakes, reservoirs) using NDWI,
   and separately calculates a Moisture Stress Index to flag where
   vegetation may be experiencing drought stress — even in areas that
   still look green in a normal photo. This is a genuinely high-impact,
   real-world technique used by drought-monitoring agencies worldwide.

   TWO INDICES IN THIS SCRIPT
   1. NDWI — finds open water (rivers, lakes, reservoirs)
   2. NDMI — finds vegetation moisture stress (early drought warning)
   ════════════════════════════════════════════════════════════════════════ */


/* STEP 1 — Choose your area */
var aoi = ee.Geometry.Rectangle([35.85, 33.55, 36.30, 33.90]);   // Bekaa Valley
// To draw your own polygon: use the tool top-left of the map, then:
// var aoi = geometry;

Map.centerObject(aoi, 10);


/* STEP 2 — Load a cloud-free composite */
var YEAR = 2023;
var START = YEAR + '-07-01';   // ✏️ summer is best for drought analysis
var END   = YEAR + '-08-31';

var image = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
  .filterDate(START, END)
  .filterBounds(aoi)
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 15))
  .median()
  .divide(10000)
  .clip(aoi);

print('✅ STEP 2 complete — composite loaded for', START, 'to', END);


/* ──────────────────────────────────────────────────────────────────────
   STEP 3 — NDWI: FIND OPEN WATER
   ──────────────────────────────────────────────────────────────────────
   NDWI = (Green − Near-Infrared) / (Green + Near-Infrared)
   Water strongly absorbs Near-Infrared light, so open water bodies
   produce HIGH NDWI values, while land produces low or negative values.
   NDWI > 0.2 is a reliable threshold for "this pixel is water."
   ────────────────────────────────────────────────────────────────────── */
var ndwi = image.normalizedDifference(['B3', 'B8']).rename('NDWI');
var waterMask = ndwi.gt(0.2);   // true/false image: is this pixel water?

var ndwiVis = {min: -0.3, max: 0.5, palette: ['white', 'lightblue', 'blue', 'darkblue']};
Map.addLayer(ndwi, ndwiVis, '1. NDWI (blue = water)');
Map.addLayer(waterMask.selfMask(), {palette: ['00BFFF']}, '2. Water Bodies Only (mask)');

print('✅ STEP 3 complete — water layers added.');


/* ──────────────────────────────────────────────────────────────────────
   STEP 4 — NDMI: VEGETATION MOISTURE STRESS (early drought warning)
   ──────────────────────────────────────────────────────────────────────
   NDMI = (Near-Infrared − SWIR1) / (Near-Infrared + SWIR1)
   This index is sensitive to the WATER CONTENT INSIDE plant leaves —
   not just whether the plant looks green. A plant can still look
   green in true color while already losing internal moisture. NDMI
   often reveals drought stress 1-3 weeks BEFORE it becomes visible
   to the naked eye or shows up clearly in NDVI.

   NDMI value reference:
     Below 0      → very dry vegetation or bare soil, high stress
     0.0 to 0.2   → moderate moisture stress
     0.2 to 0.4   → adequate moisture
     Above 0.4    → high moisture content, well-watered vegetation
   ────────────────────────────────────────────────────────────────────── */
var ndmi = image.normalizedDifference(['B8', 'B11']).rename('NDMI');

var ndmiVis = {min: -0.2, max: 0.5, palette: ['8B4513', 'FFD700', 'ADFF2F', '006400']};
Map.addLayer(ndmi, ndmiVis, '3. NDMI - Vegetation Moisture (brown=dry, green=moist)');

print('✅ STEP 4 complete — moisture stress layer added.');


/* ──────────────────────────────────────────────────────────────────────
   STEP 5 — COMPARE NDVI vs NDMI (find "hidden stress" zones)
   ──────────────────────────────────────────────────────────────────────
   This is the most valuable output of this script: areas that look
   HEALTHY in NDVI (still green) but show LOW moisture in NDMI are
   the earliest possible drought warning — vegetation that hasn't
   visibly wilted yet but is already water-stressed internally.
   ────────────────────────────────────────────────────────────────────── */
var ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI');

// "Hidden stress" = still green (NDVI > 0.3) but low moisture (NDMI < 0.1)
var hiddenStress = ndvi.gt(0.3).and(ndmi.lt(0.1));
Map.addLayer(hiddenStress.selfMask(), {palette: ['FF00FF']}, '4. HIDDEN STRESS ZONES (magenta)');

print('✅ STEP 5 complete — hidden stress zones flagged in magenta.');
print('   These areas look green/healthy but are already losing internal moisture —');
print('   in the field, this is exactly the early-warning signal irrigation teams look for.');


/* STEP 6 — Quantify results */
var waterAreaM2 = waterMask.multiply(ee.Image.pixelArea())
  .reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 30, maxPixels: 1e9});
var stressAreaM2 = hiddenStress.multiply(ee.Image.pixelArea())
  .reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 30, maxPixels: 1e9});
var totalAreaM2 = aoi.area();
var meanNDMI = ndmi.reduceRegion({reducer: ee.Reducer.mean(), geometry: aoi, scale: 100, maxPixels: 1e9});

print('═══ YOUR RESULTS (copy into your notebook) ═══');
meanNDMI.get('NDMI').evaluate(function(v) { print('Mean NDMI (moisture): ' + v.toFixed(3)); });

waterAreaM2.get('NDWI').evaluate(function(wM2) {
  totalAreaM2.evaluate(function(tM2) {
    print('Water coverage: ' + (wM2 / 1e6).toFixed(2) + ' km² (' + (wM2/tM2*100).toFixed(2) + '% of area)');
  });
});
stressAreaM2.get('NDMI').evaluate(function(sM2) {
  totalAreaM2.evaluate(function(tM2) {
    print('Hidden stress zone coverage: ' + (sM2 / 1e6).toFixed(2) + ' km² (' + (sM2/tM2*100).toFixed(2) + '% of area)');
  });
});


/* ════════════════════════════════════════════════════════════════════════
   REFLECTION PROMPT (answer this in your notebook)
   Did your area show any "Hidden Stress Zones" (Layer 4, magenta)?
   If a farmer's field fell into this category, what would you recommend
   they do — and how urgently? Compare your NDMI value to the reference
   scale above: where does your area fall, and what does that suggest
   about water availability in this particular season?
   ════════════════════════════════════════════════════════════════════════ */
'''
print("Script 4 loaded. Run the display cell below to copy it.")
````

### Script 5 — Urban vs Farmland Classifier

````python
script_5 = '''/* ════════════════════════════════════════════════════════════════════════
   SCRIPT 5 of 5 — URBAN GROWTH vs FARMLAND: Built-Up Index Mapper
   AI for Sustainable Development · Lesson 6 · ESDU @ AUB · karianet.org
   ════════════════════════════════════════════════════════════════════════
   WHAT THIS SCRIPT DOES
   Classifies your area into THREE categories — Vegetation, Built-up
   (buildings/roads/concrete), and Bare Soil/Water — using two indices
   compared against each other. Produces a single, clean classified
   map that visually separates a city from its surrounding farmland.
   This is the same fundamental technique used in urban planning and
   agricultural land-loss monitoring studies worldwide.
   ════════════════════════════════════════════════════════════════════════ */


/* STEP 1 — Choose your area (urban-rural boundary areas work especially well) */
var aoi = ee.Geometry.Rectangle([35.45, 33.80, 35.65, 33.95]);   // Beirut + surrounding area
// To draw your own polygon: use the tool top-left of the map, then:
// var aoi = geometry;

Map.centerObject(aoi, 11);


/* STEP 2 — Load a cloud-free composite */
var YEAR = 2023;
var START = YEAR + '-06-01';
var END   = YEAR + '-08-31';

var image = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
  .filterDate(START, END)
  .filterBounds(aoi)
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 15))
  .median()
  .divide(10000)
  .clip(aoi);

print('✅ STEP 2 complete — composite loaded.');


/* ──────────────────────────────────────────────────────────────────────
   STEP 3 — CALCULATE NDVI (vegetation) AND NDBI (built-up)
   ──────────────────────────────────────────────────────────────────────
   NDVI = (NIR − Red) / (NIR + Red)               → high = vegetation
   NDBI = (SWIR1 − NIR) / (SWIR1 + NIR)            → high = built-up

   Why NDBI works: buildings, roads, and concrete reflect SWIR
   (short-wave infrared) strongly but reflect NIR weakly — the
   opposite pattern from vegetation. This makes NDBI a reliable
   way to isolate human-made surfaces from natural land cover.
   ────────────────────────────────────────────────────────────────────── */
var ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI');
var ndbi = image.normalizedDifference(['B11', 'B8']).rename('NDBI');

Map.addLayer(ndvi, {min:-0.2, max:0.8, palette:['white','yellow','green']}, '1. NDVI (vegetation)', false);
Map.addLayer(ndbi, {min:-0.3, max:0.3, palette:['blue','white','red']}, '2. NDBI (built-up)', false);

print('✅ STEP 3 complete — NDVI and NDBI calculated.');


/* ──────────────────────────────────────────────────────────────────────
   STEP 4 — CLASSIFY INTO 3 CATEGORIES
   ──────────────────────────────────────────────────────────────────────
   Simple rule-based classification (no machine learning needed —
   this is "expert system" style classification using thresholds):

     Class 1 — VEGETATION   : NDVI > 0.3
     Class 2 — BUILT-UP     : NDBI > 0  AND NDVI <= 0.3
     Class 3 — BARE/WATER   : everything else
   ────────────────────────────────────────────────────────────────────── */
var vegetation = ndvi.gt(0.3);
var builtUp    = ndbi.gt(0).and(ndvi.lte(0.3));
var bareWater  = vegetation.not().and(builtUp.not());

// Combine into a single classified image: 1=vegetation, 2=built-up, 3=bare/water
var classified = ee.Image(0)
  .where(vegetation, 1)
  .where(builtUp, 2)
  .where(bareWater, 3)
  .rename('class');

var classVis = {min: 1, max: 3, palette: ['1a9850', 'd73027', 'fee08b']};
// green = vegetation, red = built-up, tan = bare/water

Map.addLayer(classified, classVis, '3. CLASSIFIED MAP (green=veg, red=urban, tan=bare)', true);

print('✅ STEP 4 complete — classified map added (visible by default).');
print('   Green = vegetation/farmland | Red = buildings/roads | Tan = bare soil or water');


/* STEP 5 — Quantify each class as a percentage of total area */
var pixelArea = ee.Image.pixelArea();

var vegArea   = vegetation.multiply(pixelArea).reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 30, maxPixels: 1e9});
var urbanArea = builtUp.multiply(pixelArea).reduceRegion({reducer: ee.Reducer.sum(), geometry: aoi, scale: 30, maxPixels: 1e9});
var totalAreaM2 = aoi.area();

print('═══ YOUR RESULTS (copy into your notebook) ═══');
vegArea.get('NDVI').evaluate(function(vM2) {
  totalAreaM2.evaluate(function(tM2) {
    print('Vegetation/farmland: ' + (vM2/1e6).toFixed(2) + ' km² (' + (vM2/tM2*100).toFixed(1) + '%)');
  });
});
urbanArea.get('NDBI').evaluate(function(uM2) {
  totalAreaM2.evaluate(function(tM2) {
    print('Built-up area: ' + (uM2/1e6).toFixed(2) + ' km² (' + (uM2/tM2*100).toFixed(1) + '%)');
  });
});


/* ════════════════════════════════════════════════════════════════════════
   REFLECTION PROMPT (answer this in your notebook)
   What percentage of your chosen area is built-up vs. vegetated?
   If you ran this same script for the same area 10 years ago, what
   direction do you predict the percentages would have shifted, and
   why? What risk does urban expansion into farmland pose for regional
   food security — and what would a city planner need from this kind
   of map to make a better-informed decision?
   ════════════════════════════════════════════════════════════════════════ */
'''
print("Script 5 loaded. Run the display cell below to copy it.")
````
